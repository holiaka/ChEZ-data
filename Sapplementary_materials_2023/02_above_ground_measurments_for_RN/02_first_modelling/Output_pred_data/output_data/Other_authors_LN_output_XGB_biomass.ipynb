```python
# Import main libraries

import numpy as np
import pandas as pd
```

# Selecting working columns for the XGBoost algorithms

```python
# We select independent variables and predicting parameter 
X_ln_Cs_geo = select_LN_Cs_with_geo
X_ln_Cs = select_LN_Cs_without_geo
X_ln_Sr_geo_D = select_LN_Sr_with_geo_D
X_ln_Sr_geo_ln_D = select_LN_Sr_with_geo_ln_D
X_ln_Sr_D = select_LN_Sr_without_geo_D
X_ln_Sr_ln_D = select_LN_Sr_without_geo_ln_D
X_ln_Tag_Cs = select_LN_Tag_Cs
X_ln_Tag_Sr = select_LN_Tag_Sr
X_Tag_Cs = select_Tag_Cs
X_Tag_Sr = select_Tag_Sr
```

```python
# Columns in "X" data collections
print(X_ln_Cs_geo.head())
print(X_ln_Cs.head())
print(X_ln_Sr_geo_D.head())
print(X_ln_Sr_geo_ln_D.head())
print(X_ln_Sr_D.head())
print(X_ln_Sr_ln_D.head())
print(X_ln_Tag_Cs.head())
print(X_ln_Tag_Sr.head())
print(X_Tag_Cs.head())
print(X_Tag_Sr.head())
```

# Load XGBoost module and learned model for predict of species and other lands

```python
#  Importing the main library for building model and its analysis

import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# Load learned models
xgb_model_LN_Cs_with_geo = xgb.XGBRegressor()
xgb_model_LN_Cs_without_geo = xgb.XGBRegressor()
xgb_model_LN_Sr_with_geo_D = xgb.XGBRegressor()
xgb_model_LN_Sr_with_geo_ln_D = xgb.XGBRegressor()
xgb_model_LN_Sr_without_geo_D = xgb.XGBRegressor()
xgb_model_LN_Sr_without_geo_ln_D = xgb.XGBRegressor()
xgb_model_LN_Tag_Cs = xgb.XGBRegressor()
xgb_model_LN_Tag_Sr = xgb.XGBRegressor()
xgb_model_Tag_Cs = xgb.XGBRegressor()
xgb_model_Tag_Sr = xgb.XGBRegressor()

xgb_model_LN_Cs_with_geo.load_model('/home/dima/Desktop/JupyterLab/GIS and ML algoritms/4_RN_content_in_wood/Output_pred_data/xgb_model_LN_Cs_with_geo.json')
xgb_model_LN_Cs_without_geo.load_model('/home/dima/Desktop/JupyterLab/GIS and ML algoritms/4_RN_content_in_wood/Output_pred_data/xgb_model_LN_Cs_without_geo.json')
xgb_model_LN_Sr_with_geo_D.load_model('/home/dima/Desktop/JupyterLab/GIS and ML algoritms/4_RN_content_in_wood/Output_pred_data/xgb_model_LN_Sr_with_geo(D).json')
xgb_model_LN_Sr_with_geo_ln_D.load_model('/home/dima/Desktop/JupyterLab/GIS and ML algoritms/4_RN_content_in_wood/Output_pred_data/xgb_model_LN_Sr_with_geo(ln_D).json')
xgb_model_LN_Sr_without_geo_D.load_model('/home/dima/Desktop/JupyterLab/GIS and ML algoritms/4_RN_content_in_wood/Output_pred_data/xgb_model_LN_Sr_without_geo(D).json')
xgb_model_LN_Sr_without_geo_ln_D.load_model('/home/dima/Desktop/JupyterLab/GIS and ML algoritms/4_RN_content_in_wood/Output_pred_data/xgb_model_LN_Sr_without_geo(ln_D).json')
xgb_model_LN_Tag_Cs.load_model('/home/dima/Desktop/JupyterLab/GIS and ML algoritms/4_RN_content_in_wood/Output_pred_data/xgb_model_LN_Tag_Cs_ver2.json')
xgb_model_LN_Tag_Sr.load_model('/home/dima/Desktop/JupyterLab/GIS and ML algoritms/4_RN_content_in_wood/Output_pred_data/xgb_model_LN_Tag_Sr_ver2.json')
xgb_model_Tag_Cs.load_model('/home/dima/Desktop/JupyterLab/GIS and ML algoritms/4_RN_content_in_wood/Output_pred_data/xgb_model_Tag_Cs_ver2.json')
xgb_model_Tag_Sr.load_model('/home/dima/Desktop/JupyterLab/GIS and ML algoritms/4_RN_content_in_wood/Output_pred_data/xgb_model_Tag_Sr_ver2.json')
```

```python
# Show all parameters of XGBoost models

print(xgb_model_LN_Cs_with_geo)
print(xgb_model_LN_Cs_without_geo)
print(xgb_model_LN_Sr_with_geo_D)
print(xgb_model_LN_Sr_with_geo_ln_D)
print(xgb_model_LN_Sr_without_geo_D)
print(xgb_model_LN_Sr_without_geo_ln_D)
print(xgb_model_LN_Tag_Cs)
print(xgb_model_LN_Tag_Sr)
print(xgb_model_Tag_Cs)
print(xgb_model_Tag_Sr)
```

```python
# Create predict values

pred_LN_Cs_geo = xgb_model_LN_Cs_with_geo.predict(X_ln_Cs_geo, ntree_limit=xgb_model_LN_Cs_with_geo.best_ntree_limit)
pred_LN_Cs = xgb_model_LN_Cs_without_geo.predict(X_ln_Cs, ntree_limit=xgb_model_LN_Cs_without_geo.best_ntree_limit)
pred_LN_Sr_geo_D = xgb_model_LN_Sr_with_geo_D.predict(X_ln_Sr_geo_D, ntree_limit=xgb_model_LN_Sr_with_geo_D.best_ntree_limit)
pred_LN_Sr_geo_ln_D = xgb_model_LN_Sr_with_geo_ln_D.predict(X_ln_Sr_geo_ln_D, ntree_limit=xgb_model_LN_Sr_with_geo_ln_D.best_ntree_limit)
pred_LN_Sr_D = xgb_model_LN_Sr_without_geo_D.predict(X_ln_Sr_D, ntree_limit=xgb_model_LN_Sr_without_geo_D.best_ntree_limit)
pred_LN_Sr_ln_D = xgb_model_LN_Sr_without_geo_ln_D.predict(X_ln_Sr_ln_D, ntree_limit=xgb_model_LN_Sr_without_geo_ln_D.best_ntree_limit)
pred_LN_Tag_Cs = xgb_model_LN_Tag_Cs.predict(X_ln_Tag_Cs, ntree_limit=xgb_model_LN_Tag_Cs.best_ntree_limit)
pred_LN_Tag_Sr = xgb_model_LN_Tag_Sr.predict(X_ln_Tag_Sr, ntree_limit=xgb_model_LN_Tag_Sr.best_ntree_limit)
pred_Tag_Cs = xgb_model_Tag_Cs.predict(X_Tag_Cs, ntree_limit=xgb_model_Tag_Cs.best_ntree_limit)
pred_Tag_Sr = xgb_model_Tag_Sr.predict(X_Tag_Sr, ntree_limit=xgb_model_Tag_Sr.best_ntree_limit)
```

```python
# Convert predicted parameters to real values

T_ln_Cs_geo = np.exp(pred_LN_Cs_geo)
T_ln_Cs = np.exp(pred_LN_Cs)
T_ln_Sr_geo_D = np.exp(pred_LN_Sr_geo_D)    
T_ln_Sr_geo_ln_D = np.exp(pred_LN_Sr_geo_ln_D)
T_ln_Sr_D = np.exp(pred_LN_Sr_D)
T_ln_Sr_ln_D = np.exp(pred_LN_Sr_ln_D)
T_ln_Tag_Cs = np.exp(pred_LN_Tag_Cs)
T_ln_Tag_Sr = np.exp(pred_LN_Tag_Sr)
```

```python
data = {'T_lnCs_geo':T_ln_Cs_geo, 'T_lnCs':T_ln_Cs, 'T_lnSr_geo_D':T_ln_Sr_geo_D, 'T_lnSr_geo_lnD':T_ln_Sr_geo_ln_D, 'T_lnSr_D':T_ln_Sr_D,
        'T_lnSr_lnD':T_ln_Sr_ln_D, 'T_lnTag_Cs':T_ln_Tag_Cs, 'T_lnTag_Sr':T_ln_Tag_Sr, 'pred_Tag_Cs':pred_Tag_Cs, 'pred_Tag_Sr':pred_Tag_Sr}
```

```python
# Add predicted values to the working dataframe
pred = pd.DataFrame(data)
```
